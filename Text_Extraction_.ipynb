```python
import PyPDF2
def extract_text(pdf1):
    with open(pdf1, 'rb') as pdf_file:
        pdf_reader = PyPDF2.PdfReader(pdf_file)
        text = ""
        for page_num in range(len(pdf_reader.pages)):
            page = pdf_reader.pages[page_num]
            text += page.extract_text() + '\n'
    return text
```

```python
pdf1 = 'A_Brief_Introduction_To_AI.pdf'
text = extract_text(pdf1)
print(text)
with open('output.txt', 'w') as text_file:
    text_file.write(text)
```

```python
import re
import PyPDF2
from PyPDF2 import PdfReader
def extract_numbers(pdf1):
    reader = PdfReader(pdf1)
    numbers = []
    for page in reader.pages:
        text = page.extract_text()
        if text:
            numbers1 = re.findall(r'\b\d+\.?\d*\b', text)
            for number in numbers1:
                    if '.' in number:
                        numbers.append(float(number))
                    else:
                        numbers.append(int(number))
                    continue
    return numbers
pdf1 = 'A_Brief_Introduction_To_AI.pdf'
numbers = extract_numbers(pdf1)
print(numbers)
```
